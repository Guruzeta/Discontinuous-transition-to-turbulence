# Simulations for tricritical DP with space: turbulent fraction $\rho$ vs space (x) for at different times

## Summary

- To simulate our mean field $\rho(x)$ vs $x$ profile we use the equations

$$ \partial_t \rho(x,t) = \tau_{p,q} \rho(x,t) - g_{p,q} \rho(x,t)^2 - c_{p,q} \rho(x,t)^3 +\Gamma \nabla^2 \rho(x,t) $$

- The outplot csv files titled "turbulent_fraction_in_space_plot_data_1000.csv" and "turbulent_fraction_in_space_plot_data_100.csv" can be found in the repository. They were generated using the mathematica file titled "TDP_with_space-turbulent fraction_at_different_distances.nb".

- The output for $L=100$ and $L=1000$ are in the files "turbulent_fraction_in_space_plot_100.csv" and "turbulent_fraction_in_space_plot_1000.csv" respectively.

- This python notebook generates plots using the aforementioned CSV files. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
data = pd.read_csv("/Users/gurukalyanjayasingh/Downloads/turbulent_fraction_in_space_plot_data_100.csv", header=None)

# Assuming your data format is correct and
labellist= [0.04,0.08,0.16, 0.32,  0.995]
```

```python


x_values = data[0] #space values for l=100
y_values = data[1] #space values for l=1000

# Plotting the data
plt.figure(figsize=(10, 9))
#plt.scatter(x_values, y_values, label='Data points', color='b')

for i in range(5):
    plt.plot(x_values[50*i:50*(i+1)], y_values[50*i:50*(i+1)])
    plt.scatter(x_values[50*i:50*(i+1)], y_values[50*i:50*(i+1)], label=r'x = %g*L'%labellist[i])


# Customize the plot
#plt.title(r'Turbulent fraction $\rho$ vs $p$ for different x values, $\Gamma/\sqrt{L}=100$',size=20)
plt.xlabel(r'p',size=40)
plt.ylabel(r'Turbulent fraction $\rho$',size=40)
plt.legend(fontsize=35)
# Increase the size of axis numbers
plt.tick_params(axis='both', which='major', labelsize=25)  # 'major' affects major ticks

# Show the plot
plt.grid(True)
plt.savefig('L_100_spatial_plot')
plt.show()
```
